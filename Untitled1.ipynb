```python
#!pip install pgeocode
import pandas as pd
import pgeocode
```

```python
 nomi=pgeocode.Nominatim('US')
```

```python
df=pd.read_excel('FFLs.xlsm', converters={'Zip': lambda x: str(x)})
df
```

```python
def lat_maker(x):
    return nomi.query_postal_code(x)[9]

    
def long_maker(x):
    return nomi.query_postal_code(x)[10] 


def location_maker(path='FFLs.xlsm',column='Zip'):
   
    df=pd.read_excel(path, converters={'Zip': lambda x: str(x)})
    df['Lattitude']=df[column].apply(lat_maker)
    df['Longitude']=df[column].apply(long_maker)
    df.to_csv("output.csv",index=False)
    return df
```

```python
df=location_maker()
```
